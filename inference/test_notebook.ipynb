```python
import random
from inference.spacy_extractor import SpacyAnchorExtractor, RequiredWordsChecker

GENERIC_NOUNS = {
        "thing", "things", "stuff", "someone", "anyone", "everyone",
        "person", "people", "man", "men", "woman", "women",
        "time", "day", "week", "month", "year",
        "place", "home", "house", "room",
        "job", "work", "boss", "company",
        "way", "kind", "sort", "part", "case", "point", "problem", "idea",
        "joke", "story",
    }

STOPWORDS = {
    # add your notebook stopwords here
    "yeah", "okay", "ok", "lol",
}

extractor = SpacyAnchorExtractor(
    generic_nouns={x.lower() for x in GENERIC_NOUNS},
    extra_stopwords={x.lower() for x in STOPWORDS},
    prefer_one_from_each=True,
    max_tries_per_pair=40,
)

checker = RequiredWordsChecker()

setup = "I bought an iPhone charger."
punch = "Now my iPhone's battery lasts longer than my patience."
pair = extractor.pick_two_anchors(setup, punch, rng=random.Random(42))
print("Picked:", pair)

print("Contains required:", checker.required_words_present(punch, "iPhone", "charger"))
```
