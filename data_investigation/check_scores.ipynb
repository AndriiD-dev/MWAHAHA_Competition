```python
import pandas as pd
import numpy as np
import sns
from plotnine import *
```

```python
dad_jokes = pd.read_csv('dad_jokes_splitted_final.csv')
one_million_jokes = pd.read_csv('one_million_jokes_cleaned_llm_poc.csv')
```

```python
dad_jokes['score'].value_counts()
```

```python
p = (
    ggplot(dad_jokes, aes(x='score'))
    + geom_histogram()
    + scale_x_log10()
)

p.show()
```

```python
one_million_jokes = one_million_jokes.dropna(how="any").copy()
one_million_jokes.isna().sum()
```

```python
one_million_jokes['score'] = one_million_jokes['score'].astype(int)
one_million_jokes['score'].unique()
```

```python
jokes = one_million_jokes[one_million_jokes["score"] >= 5].copy()

p = (
    ggplot(jokes, aes(x="score"))
    + geom_histogram()
    + scale_x_log10()
)

p.show()
```
